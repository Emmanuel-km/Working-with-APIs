## Practising CRUD operations on JSONPlaceholder api

```python
# importing the requests library
import requests
```

```python
# variable to store the base URL
BASE_URL = "https://jsonplaceholder.typicode.com"
```

```python
#C-Creating a new resource via POST request
new_post = {
    "title": "Ambiguity in APIs",
    "body": "data should be clear",
    "userId": 1
}
response = requests.post(f"{BASE_URL}/posts", json=new_post)
created_post = response.json()
print(f"Created post with ID: {created_post['id']}")
```

```python
#R-reading data from an API endpoint
response = requests.get(f"{BASE_URL}/posts")
posts = response.json()
print(f"Found {len(posts)} posts")
```

```python
#U-Updating an existing resource via PUT request
updated_data = {"title": "How should APIs be designed?"}
response = requests.put(f"{BASE_URL}/posts/1", json=updated_data)
```

```python
#D-e-Deleting a resource via DELETE request
new_post2 = {
    "title": "My New Post",
    "body": "This is the content",
    "userId": 2
}
response = requests.post(f"{BASE_URL}/posts", json=new_post2)
created_post = response.json()
print(f"Created post with ID: {created_post['id']}")
#NOW DELETING THE POST WE JUST CREATED
response = requests.delete(f"{BASE_URL}/posts/2")
print(f"Delete status: {response.status_code}")
```
